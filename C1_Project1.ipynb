```python
#FInding out the number of unique characters in the conversation.

got = []

conv = open('D:/Project/conv_problem_statement.txt','r').readlines()

for para in conv:
    characters = para.split(' ')
    if characters[0] != '\n':
        if characters[0].endswith(':') == 1:
            got.append(characters[0].replace(":",""))
        else:
            got_1 = (characters[0]+' '+characters[1])
            got.append(got_1.replace(":",""))
            
gots = set(got)

print('Number of unique characters involved in the conversation -',len(gots),"\n\nThe Character's Names are-\n")

for i in gots:
    print(i)
```

```python
#Creating a new text file for every character and saving it in a new file.

for name in gots:
    print('\n',name,":-\n")
    person = []
    names = name + ".txt"
    file = open('D:/Project/'+names,'w+')
    for para in conv:
        if para.startswith(name+":"):
            paras = para.replace(name+":"," ")
            paras = para.replace("\n","")
            u_words = paras.split(" ")
            for words in u_words:
                person.append(words)
        else:
            continue
        
    got_word = set(person)
    print(got_word)
    for n in got_word:
        if n == name:
            continue
        else:
            a = n+'\n'
            file.writelines(a)
```
